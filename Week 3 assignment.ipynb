## Importing the dataset and cleaning it

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

## Removing the rows that have Not assigned value under column: Borough

```python
df = df[df.Borough != 'Not assigned']
```

## Check if there are any 'Not assigned' values in Neighborhood

```python
print(df[df['Neighborhood']=='Not assigned'].index.values)
```

```python
df=df.sort_values(['Postal Code'],ascending=True)
df.reset_index(inplace = True, drop = True)
```
